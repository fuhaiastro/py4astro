## How to Make a Good-Looking Plot with Matplotlib

_Note that this notebook will only run properly when using the unaltered configuration files for iPython (`.ipython/profile_default/ipython_config.py`) and matplotlib (`site-packages/matplotlib/mpl-data/matplotlibrc`)_

In this notebook we explore different ways to make publication-quality figures. We will explore two types of settings:
- `InlineBackend.figure_format`
- `matplotlib.rcParams`

Note that the rcParams can also be set by customizing the `matplotlibrc` [configuration file](https://matplotlib.org/stable/tutorials/introductory/customizing.html#customizing-with-matplotlibrc-files) so that you don't need to change these values everytime you make a plot.

To get started, let's make a plot using all default plotting parameters. You'll notice the following problems in this plot:
- small fonts
- small figure size
- thin lines
- outward-pointing tickmarks
- no tickmarks on the top and the right axes

```python
import matplotlib.pyplot as plt
import numpy as np

# evaluate x,y arrays
x = np.linspace(0,360,100) # degrees
d2r = np.pi/180 # degree to radian
sinx = np.sin(x*d2r)
cosx = np.cos(x*d2r)

# plot two curves: red solid and blue dashed lines
plt.plot(x, sinx, 'r', label='sin(x)' )
plt.plot(x, cosx, 'b--', label='cos(x)' )

# set plotting ranges
plt.xlim(np.min(x), np.max(x))
plt.ylim(-1.3, 1.3)

# add legend, axis label and title
plt.legend()
plt.xlabel('x (Angle in Degrees)')
plt.ylabel('y [sin(x) or cos(x)]')
plt.title('Example Default Plot');
```

Now let's see what happens when we change `InlineBackend.figure_format` from the default `png` to `retina`. You'll notice that this single command already made the figure look much nicer with increased figure size, font size, and line thickness.

```python
%config InlineBackend.figure_format = "retina"
# Default: set([png])
# A set of figure formats to enable: png, retina, jpeg, svg, pdf.

# plot two curves: red solid and blue dashed lines
plt.plot(x, sinx, 'r', label='sin(x)' )
plt.plot(x, cosx, 'b--', label='cos(x)' )

# set plotting ranges
plt.xlim(np.min(x), np.max(x))
plt.ylim(-1.3, 1.3)

# add legend, axis label and title
plt.legend()
plt.xlabel('x (Angle in Degrees)')
plt.ylabel('y [sin(x) or cos(x)]')
plt.title('Default Retina Plot');
```

Now, let's try to make the plot look better by adjusting the following parameters in `rcParams`:
- figure aspect ratio (5x4 aspect ratio, default is 6.4, 4.8 [a 4x3 ratio])
- font size
- tickmark styles (make them point inward and add tickmarks on the top and right axes)

You may also not like the default font set (sans-serif) and want to change it to serif. The following five generic families are available and they are controlled by sub-parameters font.serif, font.sans-serif, etc.
- 'serif' (e.g., Times),
- 'sans-serif' (e.g., Helvetica),
- 'cursive' (e.g., Zapf-Chancery),
- 'fantasy' (e.g., Western), and
- 'monospace' (e.g., Courier).

First, if you don't like retina format figures:

```python
%config InlineBackend.figure_format = "png"
from matplotlib import rcParams
rcParams["figure.figsize"] = 12.5, 10.0
rcParams["font.family"] = 'serif'
rcParams["font.size"] = 20
rcParams["xtick.direction"] = 'in'
rcParams["ytick.direction"] = 'in'
rcParams["xtick.top"] = 'True'
rcParams["ytick.right"] = 'True'

# alternatively, one can use the update function
# import matplotlib
# matplotlib.rcParams.update({'font.size': 20, 'font.family': 'serif'}) 

# evaluate x,y arrays
x = np.linspace(0,360,100) # degrees
d2r = np.pi/180 # degree to radian
sinx = np.sin(x*d2r)
cosx = np.cos(x*d2r)

# plot two curves: red solid and blue dashed lines
plt.plot(x, sinx, 'r', label='sin(x)' )
plt.plot(x, cosx, 'b--', label='cos(x)' )

# set plotting ranges
plt.xlim(np.min(x), np.max(x))
plt.ylim(-1.3, 1.3)

# add legend, axis label and title
plt.legend()
plt.xlabel('x (Angle in Degrees)')
plt.ylabel('y [sin(x) or cos(x)]')
plt.title('Example Improved PNG Plot');
```

Or if you do like to use retina format:

```python
%config InlineBackend.figure_format = "retina"
from matplotlib import rcParams
rcParams["figure.figsize"] = 6.0, 4.8
rcParams["font.family"] = 'serif'
rcParams["font.size"] = 10
rcParams["xtick.direction"] = 'in'
rcParams["ytick.direction"] = 'in'
rcParams["xtick.top"] = 'True'
rcParams["ytick.right"] = 'True'

# alternatively, one can use the update function
# import matplotlib
# matplotlib.rcParams.update({'font.size': 20, 'font.family': 'serif'}) 

# evaluate x,y arrays
x = np.linspace(0,360,100) # degrees
d2r = np.pi/180 # degree to radian
sinx = np.sin(x*d2r)
cosx = np.cos(x*d2r)

# plot two curves: red solid and blue dashed lines
plt.plot(x, sinx, 'r', label='sin(x)' )
plt.plot(x, cosx, 'b--', label='cos(x)' )

# set plotting ranges
plt.xlim(np.min(x), np.max(x))
plt.ylim(-1.3, 1.3)

# add legend, axis label and title
plt.legend()
plt.xlabel('x (Angle in Degrees)')
plt.ylabel('y [sin(x) or cos(x)]')
plt.title('Example Improved Retina Plot');
```
